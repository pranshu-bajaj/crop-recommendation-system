**Importing Libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
from keras.utils import to_categorical
from google.colab import files
```

**Loading and Processing Data**

```python
crop_data_df = pd.read_csv('Crop_recommendation.csv')
crop_data_df.head()

crop_data_df = crop_data_df.dropna()

X = np.array(crop_data_df.drop(
    columns=['label'], axis=1))
print(X)

y = pd.get_dummies(crop_data_df['label'])
print(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, shuffle=True)
```

```python
#Parameters :
NB_classes = 22 #number of outputs
NB_neurones = 30 #main number of neurones
NB_features = 7 #number of inputs
activation_func = tf.keras.activations.relu #activation function used

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func
                                                   ,input_shape=(NB_features,)),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dense(NB_neurones,activation=activation_func),
                             tf.keras.layers.Dropout(0.4),
                             tf.keras.layers.Dense(NB_classes,activation=tf.keras.activations.softmax)
])

model.compile(optimizer="adam",loss=tf.keras.losses.categorical_crossentropy, metrics=['accuracy'])
model.summary()
```

**Training the model**

```python
model.fit(x=X_train,
          y=y_train,
          epochs=500,
          validation_data=(X_test,y_test),
          verbose=1,
          shuffle=True) #Train our model
```

```python
model.save("output.h5")
```

```python
performance=model.evaluate(X_test,y_test, batch_size=32, verbose=1, steps=None, )[1] * 100
print('Final accuracy : ', round(performance), '%')
```
